# Now navigate

```python
first_p = soup.p
```

```python
first_p.parent.name
```

```python
for tag in first_p.parents:
    print(tag.name)
```

```python
first_a_in_p = first_p.a
```

```python
for tag in first_a_in_p.next_siblings:
    print(tag)
```
